```python
import os
import zipfile
from PIL import Image
import numpy as np

# Create a directory for dummy images
os.makedirs('temp_raw', exist_ok=True)

# Generate 20 images with inconsistent resolutions
for i in range(20):
    # Random width and height between 300 and 800
    w, h = np.random.randint(300, 800, size=2)
    img_array = np.random.randint(0, 255, (h, w, 3), dtype=np.uint8)
    img = Image.fromarray(img_array)
    img.save(f'temp_raw/img_{i+1}.jpg')

# Zip them up
with zipfile.ZipFile('input_images.zip', 'w') as zipf:
    for root, dirs, files in os.walk('temp_raw'):
        for file in files:
            zipf.write(os.path.join(root, file), file)

print("Created 'input_images.zip' with 20 inconsistent images.")
```

```python
import os
import zipfile
import pandas as pd
from PIL import Image

# 1. Extract the images
zip_path = 'input_images.zip'
extract_path = 'Extracted_Images'
output_folder = 'Formatted_Images'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# Create output folder
os.makedirs(output_folder, exist_ok=True)

data_log = []

# 2. Resize to 256x256 and 3. Log resolutions
print("Processing images...")
for filename in os.listdir(extract_path):
    if filename.endswith(('.jpg', '.png', '.jpeg')):
        file_path = os.path.join(extract_path, filename)

        with Image.open(file_path) as img:
            # Resize image
            resized_img = img.resize((256, 256))
            resized_img.save(os.path.join(output_folder, filename))

            # Record original filename and the new consistent resolution
            data_log.append({
                'filename': filename,
                'resolution': f"{resized_img.size[0]}x{resized_img.size[1]}"
            })

# Save to CSV
df = pd.DataFrame(data_log)
df.to_csv('image_resolutions.csv', index=False)

print("Processing complete!")
print("Preview of CSV:")
print(df.head())
```
