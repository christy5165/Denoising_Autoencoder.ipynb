```python
# 1. Install necessary libraries for GenAI and Sentiment Analysis
!pip install -q transformers torch textblob

import torch
from transformers import pipeline
from textblob import TextBlob

# 2. Load the GPT Model for Summarization and Sentiment
# We use 'distilgpt2' for speed and 'distilbert' for sentiment
summarizer = pipeline("text-generation", model="distilgpt2")
sentiment_analyzer = pipeline("sentiment-analysis")

# 3. Load the text file (Simulating the upload for Colab)
# In Colab, you would upload reviews.txt to the 'files' folder
reviews = [
    "The product is amazing, build quality is top notch and I love it.",
    "Worst experience ever. The item arrived broken and customer service was rude.",
    "It is okay for the price, but there are better alternatives available."
    # ... imagine 50 reviews here
]

# 4. Processing logic
all_summaries = []
token_lengths = []

print("Processing reviews... please wait.")

for review in reviews:
    # Generate 5 alternate summaries (variations)
    # max_new_tokens controls length; num_return_sequences creates 5 variations
    outputs = summarizer(f"Summarize this: {review}", max_new_tokens=20, num_return_sequences=5, do_sample=True)

    for out in outputs:
        summary_text = out['generated_text'].replace(f"Summarize this: {review}", "").strip()
        all_summaries.append(summary_text)

        # Calculate token length (simple approximation: words = tokens)
        token_lengths.append(len(summary_text.split()))

# 5. Calculate Average Token Length
avg_token_length = sum(token_lengths) / len(token_lengths)

# 6. Perform Sentiment Analysis on the summaries
results = sentiment_analyzer(all_summaries[:10]) # Analyzing first 10 for speed

# --- OUTPUT RESULTS ---
print("-" * 30)
print(f"Total Summaries Generated: {len(all_summaries)}")
print(f"Average Token Length: {avg_token_length:.2f} tokens")
print("-" * 30)
print("Sample Sentiment Analysis on Summaries:")
for i, res in enumerate(results):
    print(f"Summary {i+1}: {res['label']} (Score: {res['score']:.2f})")

# 7. Brief Explanation (For your assignment write-up)
"""
Explanation:
The GPT model's output varies due to 'do_sample=True', which introduces randomness (temperature).
This allows the model to explore different vocabulary for the same input.
Insights: Sentiment analysis on generated summaries helps verify if the AI is
preserving the original 'emotional tone' of the user review while shortening it.
"""
```
