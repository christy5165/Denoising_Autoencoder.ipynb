```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

# 1. LOAD AND PREPROCESS DATA
print("Loading MNIST data...")
(x_train, _), (x_test, _) = tf.keras.datasets.mnist.load_data()

# Normalize and flatten (28x28 = 784 pixels)
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = x_train.reshape((len(x_train), 784))
x_test = x_test.reshape((len(x_test), 784))

# 2. DEFINE THE AUTOENCODER MODEL FUNCTION
def build_autoencoder(latent_dim):
    input_img = layers.Input(shape=(784,))
    # Encoder
    encoded = layers.Dense(128, activation='relu')(input_img)
    bottleneck = layers.Dense(latent_dim, activation='relu')(encoded)
    # Decoder
    decoded = layers.Dense(128, activation='relu')(bottleneck)
    output_img = layers.Dense(784, activation='sigmoid')(decoded)

    autoencoder = models.Model(input_img, output_img)
    autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
    return autoencoder

# 3. TRAIN BOTH MODELS
# Small Bottleneck (Size 4)
print("\nTraining Model with Latent Size 4...")
autoencoder_small = build_autoencoder(4)
autoencoder_small.fit(x_train, x_train, epochs=10, batch_size=256, verbose=0)

# Large Bottleneck (Size 64)
print("Training Model with Latent Size 64...")
autoencoder_large = build_autoencoder(64)
autoencoder_large.fit(x_train, x_train, epochs=10, batch_size=256, verbose=0)

# 4. RECONSTRUCT IMAGES
decoded_small = autoencoder_small.predict(x_test)
decoded_large = autoencoder_large.predict(x_test)

# 5. VISUALIZE RESULTS
n = 8  # Number of digits to display
plt.figure(figsize=(16, 8))

for i in range(n):
    # Original Images
    ax = plt.subplot(3, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
    plt.title("Original")
    ax.axis('off')

    # Latent Size 4 (Small)
    ax = plt.subplot(3, n, i + 1 + n)
    plt.imshow(decoded_small[i].reshape(28, 28), cmap='gray')
    plt.title("Size: 4")
    ax.axis('off')

    # Latent Size 64 (Large)
    ax = plt.subplot(3, n, i + 1 + 2 * n)
    plt.imshow(decoded_large[i].reshape(28, 28), cmap='gray')
    plt.title("Size: 64")
    ax.axis('off')

plt.tight_layout()
plt.show()

# 6. SUMMARY FOR YOUR REPORT
print("\n--- Project Summary ---")
print("1. Observations: The 'Size 4' model produces blurry reconstructions because the bottleneck is too tight.")
print("2. Observations: The 'Size 64' model produces sharp, clear images as it has more capacity to store details.")
print("3. Conclusion: Increasing the latent space size significantly improves reconstruction quality.")
```
