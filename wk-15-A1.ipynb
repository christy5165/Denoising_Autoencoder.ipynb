```python
import pandas as pd

# Create a sample dataset
data = {
    'image_url': [
        'https://images.unsplash.com/photo-1506744038136-46273834b3fb',
        'https://images.unsplash.com/photo-1470071459604-3b5ec3a7fe05',
        'https://images.unsplash.com/photo-1441974231531-c6227db76b6e',
        'https://images.unsplash.com/photo-1501785888041-af3ef285b470',
        'https://images.unsplash.com/photo-1472214103451-9374bd1c798e',
        'https://images.unsplash.com/photo-1469474968028-56623f02e42e',
        'https://images.unsplash.com/photo-1447752875215-b2761acb3c5d',
        'https://images.unsplash.com/photo-1433086566286-615b63295d20',
        'https://images.unsplash.com/photo-1518173946687-a4c8892bbd9f',
        'https://images.unsplash.com/photo-1426604966848-d7adac402bdb'
    ],
    'caption': [
        'A beautiful mountain landscape at sunset',
        'Mist rolling over green hills',
        'Sunlight filtering through a dense forest',
        'A calm lake reflecting the sky and mountains',
        'A vibrant green valley with a small stream',
        'A hiker standing on top of a rocky peak',
        'A wooden bridge crossing a forest path',
        'A majestic waterfall in a tropical jungle',
        'Close up of autumn leaves on a branch',
        'A snow-capped mountain under a clear blue sky'
    ]
}

df = pd.DataFrame(data)
df.to_csv('images_data.csv', index=False)
print("CSV file 'images_data.csv' has been created successfully!")
```

```python
import pandas as pd
import requests
import os

# --- Step 1: Load and Preview ---
df = pd.read_csv('images_data.csv')
print("Preview of the first 5 rows:")
print(df.head())

# --- Step 2: Download Images ---
save_folder = 'Raw_Images'
if not os.path.exists(save_folder):
    os.makedirs(save_folder)

local_paths = []

print("\nStarting downloads...")
for index, row in df.iterrows():
    url = row['image_url']
    file_name = f"image_{index}.jpg"
    file_path = os.path.join(save_folder, file_name)

    try:
        response = requests.get(url, stream=True)
        if response.status_code == 200:
            with open(file_path, 'wb') as f:
                f.write(response.content)
            local_paths.append(file_path)
            print(f"Downloaded: {file_name}")
        else:
            local_paths.append("Failed to download")
    except Exception as e:
        local_paths.append(f"Error: {str(e)}")

# --- Step 3: Create new CSV with local paths ---
df['local_file_path'] = local_paths
# We only want the local path and the caption for the new file
output_df = df[['local_file_path', 'caption']]
output_df.to_csv('processed_images.csv', index=False)

print("\nTask Complete: 'processed_images.csv' has been created.")
```
