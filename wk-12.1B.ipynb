```python
import pandas as pd
from transformers import pipeline
from collections import Counter
import re

# 1. Load the CSV file
# Replace 'prompts.csv' with the actual name of your uploaded file
file_name = 'prompts.csv'
df = pd.read_csv(file_name)

# Assume the column containing prompts is named 'prompt' (change if needed)
# Example: prompts = df['customer_query'].tolist()
prompts = df.iloc[:, 0].tolist() # This grabs the first column automatically

# 2. Load GPT model for text generation
generator = pipeline("text-generation", model="distilgpt2")

responses = []
print(f"Generating responses for {len(prompts)} prompts...")

for p in prompts:
    # Generate response
    output = generator(p, max_new_tokens=30, num_return_sequences=1, truncation=True)
    generated_text = output[0]['generated_text'].replace(p, "").strip()
    responses.append(generated_text)

# 3. Count Keyword Frequency
# Combine all responses into one big string to count words
all_text = " ".join(responses).lower()
words = re.findall(r'\w+', all_text) # Simple regex to get words
common_keywords = Counter(words).most_common(10)

# --- OUTPUT RESULTS ---
print("\n" + "="*40)
print("KEYWORD FREQUENCY (Top 10):")
for word, count in common_keywords:
    print(f"{word}: {count}")

print("\n" + "="*40)
print("RESPONSES FOR MANUAL EVALUATION (First 5):")
for i in range(5):
    print(f"\nPrompt {i+1}: {prompts[i]}")
    print(f"Response {i+1}: {responses[i]}")
    print("-" * 20)
    print("Score this manually for: Fluency (1-5), Relevance (1-5), Coherence (1-5)")

# 4. Summary for your Report
"""
OBSERVATIONS:
- Accuracy: Prompts that were specific (e.g., 'How do I reset my password?') usually
  resulted in more coherent outputs than vague ones.
- Behavior: The GPT model tends to repeat certain 'customer support' keywords like
  'please', 'help', and 'support' (verify this with your keyword frequency list).
- Patterns: Short prompts often led to the model hallucinating or finishing abruptly.
"""
```
