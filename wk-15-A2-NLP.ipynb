```python
# Create a messy sample .txt file with 15 captions
messy_captions = [
    "A beautiful sunset!! over the mountains...   ",
    "A cute cat @sleeping on a sofa###",
    "  Neon lights in a   cyberpunk city!!!",
    "A futuristic spaceship [landing] on Mars",
    "Vintage car driving down a coastal road???",
    "A giant robot drinking coffee...☕",
    "Underwater kingdom with colorful corals & fish!!",
    "Abstract art with swirls of blue and gold$$$",
    "A cozy cabin in the middle of a snowy forest   ",
    "Portrait of a person wearing a VR headset***",
    "  Delicious pizza with melting cheese and pepperoni!!!",
    "A mystical dragon flying through clouds!!!",
    "Ancient ruins hidden deep in the jungle@@@",
    "A futuristic library with floating books...",
    "Space station orbiting a giant gas planet!!!"
]

with open('captions.txt', 'w') as f:
    for caption in messy_captions:
        f.write(caption + '\n')

print("File 'captions.txt' created successfully!")
```

```python
import re

# 1. Load the file
with open('captions.txt', 'r') as file:
    lines = file.readlines()

cleaned_list = []

# Cleaning process
for line in lines:
    # Remove special characters (keep only alphanumeric and spaces)
    # [^a-zA-Z0-9 ] means "anything that is NOT a letter, number, or space"
    clean_text = re.sub(r'[^a-zA-Z0-9 ]', '', line)

    # Remove extra spaces and leading/trailing whitespace
    clean_text = ' '.join(clean_text.split())

    if clean_text: # Avoid empty lines
        cleaned_list.append(clean_text)

# 2. Save the cleaned captions
with open('Cleaned_Captions.txt', 'w') as f:
    for caption in cleaned_list:
        f.write(caption + '\n')

# Preview the first 5 for the assignment
print("Preview of Cleaned_Captions.txt:")
print(cleaned_list[:5])
```
