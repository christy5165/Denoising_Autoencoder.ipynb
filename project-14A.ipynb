```python
# 1. Install necessary libraries
!pip install -q openai requests

import os
import requests
from openai import OpenAI

# --- CONFIGURATION ---
# Replace 'YOUR_API_KEY_HERE' with your actual OpenAI API Key
api_key = 'AIzaSyAorjy4XmG8aQYwMU6cp55ESAzMheY23RY'
client = OpenAI(api_key=api_key)

# 2. STEP 1: Programmatically create the .txt file with 5 stylistic prompts
file_content = """red apple in watercolor
blue car in cartoon style
yellow flower in oil painting
green mountain in pixel art
purple cat in neon cyberpunk style"""

txt_filename = "style_prompts.txt"
with open(txt_filename, "w") as f:
    f.write(file_content)

print(f"-> Successfully created {txt_filename}")

# 3. STEP 1 (Cont.): Read the prompts from the file
with open(txt_filename, "r") as f:
    prompts = [line.strip() for line in f if line.strip()]

# 4. STEP 2 & 3: Generate images and save to "Conditioned_Images" folder
output_folder = "Conditioned_Images"
os.makedirs(output_folder, exist_ok=True)

print("\n-> Starting image generation...")

for i, prompt in enumerate(prompts):
    print(f"Generating ({i+1}/5): {prompt}...")
    try:
        # Request generation
        response = client.images.generate(
            model="dall-e-2",
            prompt=prompt,
            n=1,
            size="512x512"
        )

        # Download the image
        img_url = response.data[0].url
        img_data = requests.get(img_url).content

        # Create descriptive filename
        filename = prompt.replace(" ", "_").lower() + ".png"
        save_path = os.path.join(output_folder, filename)

        with open(save_path, 'wb') as handler:
            handler.write(img_data)

        print(f"   [Saved] {save_path}")

    except Exception as e:
        print(f"   [Error] Could not generate image for '{prompt}': {e}")

# 5. STEP 4: Short Note on DALL·E Style Adaptation
note = """
---
### Note: How DALL·E Adapted the Style
DALL·E uses a process called conditioning to interpret the stylistic keywords in each prompt.
- For 'watercolor', the model simulates translucent layers and bleeding edges.
- For 'oil painting', it emphasizes thick brushstrokes and texture.
- For 'pixel art', it constrains the output to a grid-like structure with a limited color palette.
By recognizing these 'style tokens', DALL·E shifts its internal weights to match the visual language
requested, ensuring the subject (apple, car, etc.) is rendered through the lens of that specific art medium.
---
"""
print(note)
```
